```python
def entity_detection(sentence):
    doc = nlp(text)
    language = doc._.language['language']
    if language == 'en':
        doc = nlp(sentence + ' .')
    elif language == 'es':
        doc = nlp_es(sentence)
    else:
        raise ValueError('El idioma debe ser "ingles" o "español".')
    entities = []
    token = ""
    k = 0
    n = len(doc)
    for index in range(n):
        #print(f"index = {index} token = {doc[index].text} {doc[index].pos_}")
        if k>0:
            k-=1
            continue
        else:
            if doc[index].pos_ in ['VERB']:
                token = doc[index].text
                entities.append(token)
            elif doc[index].pos_ == 'NOUN':
                if index < n - 1 and doc[index + 1].pos_ == 'NUM':
                  token = doc[index].text + ' ' + doc[index + 1].text
                  k = 1
                else:
                  token = doc[index].text
                entities.append(token)
            elif index < n - 1 and doc[index].pos_ == 'NUM' and doc[index + 1].pos_ == 'NOUN':
                token = doc[index].text + ' ' + doc[index + 1].text
                k = 1
                entities.append(token)
            elif doc[index].pos_ == 'PROPN':
                if index < n - 2 and doc[index + 1].pos_ == 'PUNCT' and doc[index + 2].pos_ == 'PROPN':
                  token = doc[index].text + doc[index + 1].text + doc[index + 2].text
                  k  = 2
                else:
                  token = doc[index].text
                entities.append(token)
    print("Entidades :", entities)
    return entities

#sentence = "For a given COVID-19 case with symptom onset on the t-th day, the number of secondary cases follows a negative binomial (NB) distribution with a mean at Reff(t) and a dispersion parameter at k"
#entities = entity_detection(sentence)
#print("Entidades :", entities)
```

```python
def save_csv2(file_path, dataset):
    with open(file_path, 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(["input", "output"])
        for data1,data2 in zip(dataset[0],dataset[1]):
            input_text = str(data1)
            output_text = str(data2)
            writer.writerow([input_text, output_text])
```

```python
def create_list(n):
    result = []
    for i in range(n):
        result.append(i)
    return result 
```

```python
def get_input_output(sentence):
    keyphrases = []
    labels = []
    for keyphrase in sentence.keyphrases:
      keyphrases.append(keyphrase.text)
      labels.append(keyphrase.label)
    return sentence.text, keyphrases, labels

def eval_model(collection):
    CORRECT, INCORRECT, PARTIAL, MISSING, SPURIOUS = 0, 0, 0, 0, 0
    input_spacy = []
    input_T5 = []
    output_T5 = []
    for sentence in collection.sentences:
      t1, t2, t3 = get_input_output(sentence)
      print(sentence)
      input_spacy.append(t1)
      input_T5.append(t2)
      output_T5.append(t3)
    rec_T5, prec_T5, f1_T5 = eval_T5(input_T5, output_T5) 
    rec_spacy,prec_spacy,f1_spacy = eval_spacy(input_spacy,input_T5)
    rec_T5_spy,prec_T5_spy,f1_T5_spy = eval_T5_with_spacy(input_spacy, input_T5, output_T5)

    rec = (rec_T5 + rec_spacy + rec_T5_spy)/3
    prec = (prec_T5 + prec_spacy + prec_T5_spy)/3
    f1 = (f1_T5 + f1_spacy + f1_T5_spy)/3
    return rec, prec, f1
```

```python
def eval_T5(input_T5, output_T5):
    CORRECT, INCORRECT = 0, 0
    results = []
    inputs = []
    print('Evaluating T5')
    for keys, labels in zip(input_T5,output_T5):
      for input, output in zip(keys,labels):
        print("input", input)
        print("output", output)
        result = model.predict(input)
        results.append(result)
        inputs.append(inputs)
        print("result", result)

        if result[0] == output:
          CORRECT += 1
        else:
          INCORRECT += 1
    print("CORRECT ",CORRECT)
    print("INCORRECT", INCORRECT)
    file_name = 'result_test_T5.csv'
    file_path = '/content/drive/MyDrive/' + file_name  # Ruta completa de Google Drive
    data = [inputs, results]
    save_csv2(file_name, data)# Guardar .csv local
    save_csv2(file_path, data )# Guardar .csv en Google Drive

    # Precision, Recobrado y la puntuación F1 de todas las palabras
    rec = float(CORRECT/(CORRECT + INCORRECT)) # prec = rec
    f1 = float(rec*rec/(rec + rec))
    print(f"T5: rec = {rec}, prec =  {rec}, f1 = {f1}")
    return rec, rec, f1
    

def eval_spacy(input_spacy, output_spacy):
    CORRECT, PARTIAL, MISSING, SPURIOUS = 0, 0, 0, 0
    counter = 0
    results = []
    inputs = []
    for sentence, keys in zip(input_spacy,output_spacy):
        counter += 1
        check = [False]*len(keys)
        print('Evaluating sentence', counter)
        result_spacy = entity_detection(sentence)
        inputs.append(sentence)
        results.append(result_spacy)

        for key_dev in result_spacy:
            for i, key_gold in enumerate(keys):
                if key_dev == key_gold and not check[i]:
                    CORRECT += 1
                    check[i] = True
                    break
        for key_dev in result_spacy:
            for i, key_gold in enumerate(keys):
                if key_dev in key_gold and not check[i]:
                    PARTIAL += 1
                    check[i] = True
                    break
                else:
                    SPURIOUS += 1
                    break

        for e in check:
            if not e:
                MISSING += 1

    file_name = 'result_test_spacy.csv'
    file_path = '/content/drive/MyDrive/' + file_name  # Ruta completa de Google Drive
    data = [inputs, results]
    save_csv2(file_name, data)# Guardar .csv local
    save_csv2(file_path, data )# Guardar .csv en Google Drive

    rec = float(CORRECT + (1/2)*PARTIAL/(CORRECT + PARTIAL + MISSING))
    prec =  float(CORRECT/(CORRECT + PARTIAL + SPURIOUS))
    f1 = float(2*(prec * rec)/(prec + rec))
    print(f"SPACY: rec = {rec}, prec =  {rec}, f1 = {f1}")
    return rec, prec, f1

def eval_T5_with_spacy(input_spacy, input_T5, output_T5):
    CORRECT, INCORRECT, PARTIAL, MISSING, SPURIOUS = 0, 0, 0, 0, 0
    counter = 0
    inputs = []
    results = []
    for sentence, keys, labels in zip(input_spacy,input_T5,output_T5):
        temp = []
        counter += 1
        check = [False]*len(keys)
        print('Evaluating sentence', counter)
        result_spacy = entity_detection(sentence)
        inputs.append(sentence)

        for key_dev in result_spacy:
            print("labels", labels)
            print("keys", keys)
            n = len(keys)
            lista = create_list(n)
            for i, key_gold,label_gold in zip(lista,keys,labels):
                if key_dev == key_gold and not check[i]:
                    label_dev = model.predict(key_gold)
                    if label_dev == label_gold:
                        CORRECT += 1
                        check[i] = True
                    else:
                        INCORRECT += 1
                        check[i] = True
                    temp.append(label_dev)
                    break
        for key_dev in result_spacy:
            for i, key_gold in enumerate(keys):
                if key_dev in key_gold and not check[i]:
                    PARTIAL += 1
                    check[i] = True
                    temp.append(key_dev)
                    break
                else:
                    SPURIOUS += 1
                    temp.append(key_dev)
                    break

        for e in check:
            if not e:
                MISSING += 1
        results.append(temp)

    file_name = 'result_test_T5_with_spacy.csv'
    file_path = '/content/drive/MyDrive/' + file_name  # Ruta completa de Google Drive
    data = [inputs, results]
    save_csv2(file_name, data)# Guardar .csv local
    save_csv2(file_path, data )# Guardar .csv en Google Drive

    rec = float(CORRECT + 1/2*PARTIAL/(CORRECT + INCORRECT + PARTIAL + MISSING))
    prec =  float(CORRECT/(CORRECT + INCORRECT + PARTIAL + SPURIOUS))
    f1 = float(2*(prec * rec)/(prec + rec))
    print(f"T5 with spacy: rec = {rec}, prec =  {rec}, f1 = {f1}")
    return rec, prec, f1


ref = Path('./datasets/test/scenario1-main')
collection = Collection().load_dir(ref)
rec_b, prec_b, f1_b = eval_model(collection)
print("REC_B: ", rec_b)
print("PREC_B: ", prec_b)
print("F1_B: ", f1_b)
```
