```python
def entity_detection(sentence):
    doc = nlp(sentence)
    language = doc._.language['language']
    if language == 'en':
        doc = nlp(sentence + ' .')
    else:# language == 'es':
        doc = nlp_es(sentence)
    entities = []
    token = ""
    k = 0
    n = len(doc)
    for index in range(n):
        #print(f"index = {index} token = {doc[index].text} {doc[index].pos_}")
        if k>0:
            k-=1
            continue
        else:
            if doc[index].pos_ in ['VERB']:
                token = doc[index].text
                entities.append(token)
            elif doc[index].pos_ == 'NOUN':
                if index < n - 1 and doc[index + 1].pos_ == 'NUM':
                  token = doc[index].text + ' ' + doc[index + 1].text
                  k = 1
                else:
                  token = doc[index].text
                entities.append(token)
            elif index < n - 1 and doc[index].pos_ == 'NUM' and doc[index + 1].pos_ == 'NOUN':
                token = doc[index].text + ' ' + doc[index + 1].text
                k = 1
                entities.append(token)
            elif doc[index].pos_ == 'PROPN':
                if index < n - 2 and doc[index + 1].pos_ == 'PUNCT' and doc[index + 2].pos_ == 'PROPN':
                  token = doc[index].text + doc[index + 1].text + doc[index + 2].text
                  k  = 2
                else:
                  token = doc[index].text
                entities.append(token)
    print("Entidades :", entities)
    return entities

#sentence = "Puede ser benigno (no canceroso) o maligno (canceroso)."
#entities = entity_detection(sentence)
```

```python
from pathlib import Path
from copy import copy
from utils.anntools import Collection
import re
from re import sub
import csv

def sum_all(lista):
    total = 0
    for item in lista:
        total+= len(item)
    return total

def generate_t5_input_output_format1(collection):
    dataset_T5 = []
    dataset_spacy = []
    for sentence in collection.sentences:
        input_spacy = sentence.text
        output_spacy = []
        for keyphrase in sentence.keyphrases:
            #making the pairs
            input_T5 = keyphrase.text
            #print("input T5: ", input_T5)
            #print(sentence.keyphrases)
            output_T5 = f"{keyphrase.label}"
            #print("output T5: ", output_T5)
            output_spacy.append(input_T5)
            dataset_T5.append((input_T5, output_T5))
        print("input spacy ", input_spacy)
        print("output spacy ", output_spacy)
        print()
        dataset_spacy.append((input_spacy, output_spacy))
    return dataset_T5, dataset_spacy


ref = Path('./datasets/train')
collection = Collection().load_dir(ref)
train_dataset,dataset_spacy  = generate_t5_input_output_format1(collection)
MAX_INPUT_TOKENS = max([len(data[0]) for data in train_dataset])
MAX_OUTPUT_TOKENS = max([len(data[1]) for data in train_dataset])
MAX_INPUT_TOKENS_SPACY = max([len(data[0]) for data in dataset_spacy])
MAX_OUTPUT_TOKENS_SPACY = max([sum_all(data[1]) for data in dataset_spacy])
```

```python
def create_list(n):
    result = []
    for i in range(n):
        result.append(i)
    return result 
```

```python
def get_input_output(sentence):
    keyphrases = []
    labels = []
    for keyphrase in sentence.keyphrases:
      keyphrases.append(keyphrase.text)
      labels.append(keyphrase.label)
    return sentence.text, keyphrases, labels

def eval_model(test_dataset_T5, test_dataset_spacy):
    rec_T5, prec_T5, f1_T5 = eval_T5(test_dataset_T5)
    rec_spacy,prec_spacy,f1_spacy = eval_spacy(test_dataset_spacy)
    rec_T5_spy,prec_T5_spy,f1_T5_spy = eval_T5_with_spacy(test_dataset_T5, test_dataset_spacy)

    rec = (rec_T5 + rec_spacy + rec_T5_spy)/3
    prec = (prec_T5 + prec_spacy + prec_T5_spy)/3
    f1 = (f1_T5 + f1_spacy + f1_T5_spy)/3
    return rec, prec, f1

def eval_T5(test_dataset_T5):
    print("ENtrando eval T5")
    CORRECT, INCORRECT = 0, 0
    results = []
    print('Evaluating T5')
    for key, label in test_dataset_T5:
      print("input key ", key)
      print("output label ", label)
      result = model.predict(key)
      #result = [label]
      print("result ", result[0])
      results.append((key, result[0]))

      if result[0] == label:
        CORRECT += 1
      else:
        INCORRECT += 1
    print("CORRECT ",CORRECT)
    print("INCORRECT", INCORRECT)
    file_name = 'result_test_T5.csv'
    file_path = '/content/drive/MyDrive/' + file_name  # Ruta completa de Google Drive
    save_csv(file_name, results)# Guardar .csv local
    save_csv(file_path, results )# Guardar .csv en Google Drive

    # Precision, Recobrado y la puntuación F1 de todas las palabras
    rec = CORRECT/(CORRECT + INCORRECT) # prec = rec
    f1 = rec*rec/(rec + rec)
    print(f"T5: rec = {rec}, prec =  {rec}, f1 = {f1}")
    return rec, rec, f1


def eval_spacy(test_dataset_spacy):
    print("ENtrando eval spacy")
    CORRECT, PARTIAL, MISSING, SPURIOUS = 0, 0, 0, 0
    counter = 0
    results = []
    for data in test_dataset_spacy[100:]: #data[0] = sentence, data[1] = keys
        counter += 1
        check_gold = [False]*len(data[1])
        print('Evaluating sentence', counter)
        result_spacy = entity_detection(data[0])
        check_dev = [False]*len(result_spacy)
        results.append((data[0], result_spacy))
        print("data = ", data[1])

        for j, key_dev in enumerate(result_spacy):
            for i, key_gold in enumerate(data[1]):
                if key_dev == key_gold and not check_gold[i]:
                    print("CORRECT = ", key_dev)
                    CORRECT += 1
                    check_gold[i] = True
                    check_dev[j] = True
                    break

        for j, key_dev in enumerate(result_spacy):
            _in = False
            if check_dev[j]: continue
            for i, key_gold in enumerate(data[1]):
                if not check_gold[i] and key_dev in key_gold:
                    print("PARTIAL = ", key_dev)
                    PARTIAL += 1
                    check_gold[i] = True
                    check_dev[j] = True
                    _in = True
                    break
            if not _in :
                print("SPURIOUS = ", key_dev)
                SPURIOUS += 1
        for i, e in enumerate(check_gold):
            if not e:
                print("MISSING = ", data[1][i])
                MISSING += 1

    file_name = 'result_test_spacy.csv'
    file_path = '/content/drive/MyDrive/' + file_name  # Ruta completa de Google Drive
    save_csv(file_name, results)# Guardar .csv local
    save_csv(file_path, results )# Guardar .csv en Google Drive

    rec = (CORRECT + (PARTIAL/2))/(CORRECT + PARTIAL + MISSING)
    prec =  (CORRECT + (PARTIAL/2))/(CORRECT + PARTIAL + SPURIOUS)
    f1 = 2*(prec * rec)/(prec + rec)
    print(f"SPACY: rec = {rec}, prec =  {rec}, f1 = {f1}")
    return rec, prec, f1

def eval_T5_with_spacy(test_dataset_T5, test_dataset_spacy):
    print("ENtrando eval T5 with spacy")
    CORRECT, INCORRECT, PARTIAL, MISSING, SPURIOUS = 0, 0, 0, 0, 0
    counter = 0
    start = 0
    end = 0
    results = []
    for data in test_dataset_spacy[100:]:  #data[0] = sentence, data[1] = keys
        n = len(data[1])
        end += n
        temp = []
        counter += 1
        check_gold = [False]*n
        print('Evaluating sentence', counter)
        result_spacy = entity_detection(data[0])
        check_dev = [False]*len(result_spacy)
        print("data = ", data[1])

        for j, key_dev in enumerate(result_spacy):
            lista = create_list(n)
            for i, data2 in enumerate(test_dataset_T5[start:end]): # data2[0] = key_gold, data2[1] = label_gold
                if key_dev == data2[0] and not check_gold[i]:
                    label_dev = model.predict(data2[0])
                    #label_dev = data2[1]
                    if label_dev == data2[1]:
                        print(f"CORRECT = {key_dev} :{label_dev}")
                        CORRECT += 1
                    else:
                        print(f"INCORRECT = {key_dev} :{label_dev}")
                        INCORRECT += 1
                    check_gold[i] = True
                    check_dev[j] = True
                    temp.append((key_dev, label_dev))
                    break
        for j, key_dev in enumerate(result_spacy):
            _in = False
            if check_dev[j]: continue
            for i, data2 in zip(lista, test_dataset_T5[start:end]): # data2[0] = key_gold, data2[1] = label_gold
                if not check_gold[i]:
                    if key_dev in data2[0]: 
                        print("PARTIAL = ", key_dev)
                        label_dev = model.predict(data2[0])
                        #label_dev = "MODEL"
                        PARTIAL += 1
                        temp.append((key_dev, label_dev))
                        check_gold[i] = True
                        check_dev[j] = True
                        _in = True
                        break
            if not _in:
                print("SPURIOUS = ", key_dev)
                label_dev = model.predict(data2[0])
                #label_dev = "MODEL"
                SPURIOUS += 1
                temp.append((key_dev, label_dev))
            

        for i, e in enumerate(check_gold):
            if not e:
                print("MISSING = ", data[1][i])
                MISSING += 1
        results.append((data[0], temp))
        start = end

    file_name = 'result_test_T5_with_spacy.csv'
    file_path = '/content/drive/MyDrive/' + file_name  # Ruta completa de Google Drive
    save_csv(file_name, results)# Guardar .csv local
    save_csv(file_path, results )# Guardar .csv en Google Drive

    rec = (CORRECT + (PARTIAL/2))/(CORRECT + INCORRECT + PARTIAL + MISSING)
    prec =  (CORRECT + (PARTIAL/2))/(CORRECT + INCORRECT + PARTIAL + SPURIOUS)
    f1 = 2*(prec * rec)/(prec + rec)
    print(f"T5 with spacy: rec = {rec}, prec =  {rec}, f1 = {f1}")
    return rec, prec, f1


rec_b, prec_b, f1_b = eval_model(test_dataset, test_dataset_spacy)
print("REC_B: ", rec_b)
print("PREC_B: ", prec_b)
print("F1_B: ", f1_b)
```
